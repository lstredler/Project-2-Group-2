```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
population_file = "2019_Census.csv"
population_df = pd.read_csv(population_file)
population_df.head()
```

```python
state_file = "state_vaccinations.csv"
state_df = pd.read_csv(state_file)
state_df.head()
```

```python
merged_df = pd.merge(state_df, population_df, on="STATE")
merged_df.head()
```

```python
final_df = merged_df.loc[merged_df["date"] == "6/22/2021"]
final_df
```

```python
final_df.count()
```

```python
final_df.to_csv(r'C:\HomeworkNU\updated_state_vaccinations2.csv', index = False)
```

```python
final_cols = ["STATE", "people_vaccinated",  "POPESTIMATE2019", "lat", "long"]
final_df_transformed = final_df[final_cols]
final_df_transformed
```

```python
final_df_transformed["percent_vaccinated"] = final_df_transformed["people_vaccinated"] / final_df_transformed["POPESTIMATE2019"]

final_df_transformed
```
